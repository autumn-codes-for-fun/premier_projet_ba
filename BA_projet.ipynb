```python
"""convertis les pourcentages en float plus facilement utilisable pour des calculs
Je crée la colonnne discount_percentage_bis pour eviter de modifier à chaque fois ma colonne discount_percentage et modifier les resultats en place"""
amazon_products['discount_percentage_bis'] = (
    amazon_products['discount_percentage']
        .astype(str)
        .str.rstrip('%')
        .astype(float)
        .div(100)
)
amazon_products['discounted_price'] = (amazon_products['discounted_price']
    .astype(str)
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
    .astype(float))
"""convertis les prix et les scores (rating) et des nombres de score (rating count)en float plus facilement utilisable pour des calculs"""
amazon_products['actual_price'] = (amazon_products['actual_price']
    .astype(str)
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
    .astype(float))
amazon_products['rating_count'] = (amazon_products['rating_count']
   .astype(str)
    .str.replace(',','',regex=False)
    .astype(float))
"""Conversion de la colonne rating en Float. Les valeurs non numériques ont été converties en NaN et exclues des calculs statistiques."""
amazon_products['rating'] = pd.to_numeric(amazon_products['rating'], errors="coerce") 
"""divise la colone category en liste en créant la nouvelle colonne category_bis, cela permettra de filtrer le tableau avec plus de precisions plus tard"""
amazon_products['category_bis'] = (amazon_products['category']
    .astype(str)
    .str.split("|"))
""" après avoir séparé la categorie en liste je recherche la longueur de liste maximale pour déterminer combien de colonne il faudra ajouter au tableau
dans ce cas cela correspond à 7 colonnes qui constitue 7 sous catégories de produitss"""
max_category_length = amazon_products['category_bis'].apply(len).max()
""" création des 7 sous categories avec la fonction apply pour reporter les valeurs d'une colonne à l'autre
+ gestion d'erreur si la liste est vide avec if len(x) > i else None """
for i in range(max_category_length):
    amazon_products[f'sub_{i+1}'] = amazon_products['category_bis'].apply(
        lambda x: x[i] if len(x) > i else None
    )
"""recherche toutes les valeurs manquante dans chaque colonne"""
amazon_products.isna().sum()

```
